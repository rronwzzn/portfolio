# PART II

## 1. Importing Pandas and Loading Data

```python
import numpy as np
import pandas as pd
```

```python
# Load CSV file into a DataFrame

df=pd.read_csv('https://raw.githubusercontent.com/renatomaaliw3/public_files/refs/heads/master/Data%20Sets/healthcare_dataset.csv')
df
```

```python
print("DataFrame Head:\n", df.head())
```

# 2. Exploring The DataFrame

```python
# Display first five rows of the DataFrame

print("First five rows:\n", df.head())
```

```python
# Check for missing values

print("Missing values:\n", df.isnull().sum())
```

```python
# Get summary statistics

print("Summary statistics:\n", df.describe())
```

# 3. Filtering and Sorting Data

```python
# Filter rows where 'age' column is greater than 30

filtered_df=df[df['Age']>30]
print("Rows where age is greater than 30:\n", filtered_df)
```

```python
# Sort DataFrame by 'age' in descending order

sorted_df=df.sort_values(by='Age',ascending=False)
print("DataFrame sorted by age:\n", sorted_df.head())
```

# 4. GroupBy and Aggregration

```python
# Group by 'Hospital' and calculate the mean of billing amount

grouped_df=df.groupby('Hospital')['Billing Amount'].mean()
print("Mean billing amount per hospital:\n", grouped_df)
```

# 5. Data Cleaning and Manipulation

```python
df['Age'].fillna(df['Age'].mean(), inplace=True)
print("DataFrame with missing values filled:\n", df)
```

```python
# Create a new column 'name_hospital' by combining 'Name' and 'Hospital'

df['name_hospital']=df['Name']+' '+df['Hospital']
print("DataFrame with new column:\n", df)
```
