# **Data Visualization with Matplotlib using Scatter Plot**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

**1. Create a scatter plot using a dataset of your choice**

```python
# Importing the chosen dataset

df = pd.read_csv('https://raw.githubusercontent.com/renatomaaliw3/public_files/refs/heads/master/Data%20Sets/scatter_plot.csv')
df.head()
```

```python
# Create a scatter plot

plt.figure(figsize=(8,4))

#Plot Variable 1 against the Variable 4
plt.scatter(df['Variable 1'], df['Variable 4'])

plt.show()
```

**2. Change the color,size, and marker style of the data points.**

```python
# Create a basic scatter plot
plt.figure(figsize=(8,4))

# Plot Variable 1 against the Variable 4 (Changing the color, size, and marker style)
plt.scatter(df['Variable 1'], df['Variable 4'], color='blue', s=50, marker='^', edgecolor='black', linewidths= 1)

# Showing the grid
plt.grid(True, color='gray')

plt.show()
```

**3. Use different colors to represent different groups or categories.**

```python
# Sorting the data frame values
sorted_df = df.sort_values('Label')

# Create a scatter plot
plt.figure(figsize=(8,4))

# Setting up colors to visually differentiate groups
true_color = ['yellow', 'blue', 'green']

# Using the color palette to visually represent each group
for index, label in enumerate(sorted_df['Label'].unique()):
  subset = sorted_df[sorted_df['Label'] == label]
  plt.scatter(subset['Variable 1'], subset['Variable 4'], label = label,
              marker = '^', s = 80, edgecolor = 'black', c=true_color[index])

# Showing the grid
plt.grid(True, color='gray')

plt.show()
```

**4. Add labels, titles, and a legend if applicable**

```python
# Sorting the data frame values
sorted_df = df.sort_values('Label')

# Create a scatter plot
plt.figure(figsize=(8,4))

# Setting up colors to visually differentiate groups
true_color = ['yellow', 'blue', 'green']

# Using the color palette to visually represent each group
for index, label in enumerate(sorted_df['Label'].unique()):
  subset = sorted_df[sorted_df['Label'] == label]
  plt.scatter(subset['Variable 1'], subset['Variable 4'], label = label,
              marker = '^', s = 80, edgecolor = 'black', c=true_color[index])

# Adding labels and title
plt.xlabel('Variable 1', fontsize=12, labelpad=10)
plt.ylabel('Variable 4', fontsize=12, labelpad=10)
plt.title('Scatter Plot of Variable 1 against Variable 4', pad=12, fontweight='bold')

# Showing the legend in best and visible position
plt.legend(loc='best')

# Showing the grid
plt.grid(True, color='gray')

plt.show()
```

**5. Experiment with transparency (alpha) to obeserve overlapping data points**

```python
# Create a scatter plot
plt.figure(figsize=(8, 4))

# Sorting the data frame values
sorted_df = df.sort_values('Label')

# Setting up colors to visually differentiate groups
true_color = ['yellow', 'blue', 'green']

# Using the color palette to visually represent each group
for idx, label in enumerate(sorted_df['Label'].unique()):
  subset = sorted_df[sorted_df['Label'] == label]
  plt.scatter(subset['Variable 1'], subset['Variable 4'], label = label,
              marker = '^', s = 50, alpha = 0.5, edgecolor = 'black', c=true_color[idx])

# Adding labels and title
plt.xlabel('Variable 1', fontsize = 13, labelpad = 10)
plt.ylabel('Variable 4', fontsize = 13, labelpad = 10)
plt.title('Scatter Plot of Variable 1 against Variable 4', pad = 10, fontweight = 'bold')

# Showing the legend in best and visible position
plt.legend(loc = 'best')

# Adding grid + transparency
plt.grid(True, alpha = 0.25, color = 'gray')

plt.show()
```
