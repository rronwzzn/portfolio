Activity 5: IF Statement and Loops

```python
  # If-elif-else

  num = int(input("Enter a number: "))

if num > 0:
    print("The number you entered is positive.")
elif num == 0:
    print("The number you entered is zero.")
else:
    print("The number you entered is negative.")
```
