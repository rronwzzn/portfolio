```python
# Imports

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.neighbors import NearestNeighbors
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/renatomaaliw3/public_files/refs/heads/master/Data%20Sets/creditcard_reduced.csv')
df
```

```python
# Select features V1 and V12
X = df[['V1', 'V12']]

# Apply DBSCAN with specified parameters
dbscan = DBSCAN(eps=31, min_samples=51)
clusters = dbscan.fit_predict(X)

# Add cluster labels to the dataframe
df['DBSCAN_Label'] = clusters

# Filter for anomalies (cluster label -1) and original class 1 (anomalous)
true_positives = df[(df['DBSCAN_Label'] == -1) & (df['Class'] == 1)]

# Count the number of true positives
num_true_positives = len(true_positives)

print(f"Number of TRUE POSITIVE (anomalous) data when V1 and V12 are paired: {num_true_positives}")
```

```python
# Filter for anomalous data points (cluster label -1)
anomalies = df[df['DBSCAN_Label'] == -1]

# Check if there are any anomalies
if not anomalies.empty:
  # Get the value of the last data point in the 'V12' column
  last_v12_anomaly = anomalies['V12'].iloc[-1]
  print(f"The value of the last data based on V12 column detected as anomalous is: {last_v12_anomaly}")
else:
  print("No anomalous data points were detected by DBSCAN.")
```

```python
# Select features V3 and V11
X_V3_V11 = df[['V3', 'V11']]

# Apply DBSCAN with specified parameters (eps=31, min_samples=51)
dbscan_V3_V11 = DBSCAN(eps=31, min_samples=51)
clusters_V3_V11 = dbscan_V3_V11.fit_predict(X_V3_V11)

# Add cluster labels to the dataframe (using a new column name to avoid overwriting)
df['DBSCAN_Label_V3_V11'] = clusters_V3_V11

# Filter for anomalies (cluster label -1) and original class 1 (anomalous)
true_positives_V3_V11 = df[(df['DBSCAN_Label_V3_V11'] == -1) & (df['Class'] == 1)]

# Count the number of true positives
num_true_positives_V3_V11 = len(true_positives_V3_V11)

print(f"Number of TRUE POSITIVE (anomalous) data when V3 and V11 are paired: {num_true_positives_V3_V11}")
```

```python
# Filter for anomalous data points (cluster label -1) using the V3_V11 clustering results
anomalies_V3_V11 = df[df['DBSCAN_Label_V3_V11'] == -1]

# Check if there are any anomalies
if not anomalies_V3_V11.empty:
  # Get the value of the last data point in the 'V11' column
  last_v11_anomaly = anomalies_V3_V11['V11'].iloc[-1]
  print(f"The value of the last data based on V11 column detected as anomalous when V3 and V11 are paired is: {last_v11_anomaly}")
else:
  print("No anomalous data points were detected by DBSCAN when V3 and V11 are paired.")
```

K-Means partitions data into a predefined number of clusters based on centroid distance, while DBSCAN groups together data points that are closely packed, identifying outliers as noise.

Setting epsilon and min_samples is challenging because these values are highly dependent on the data's density and scale, and incorrect values can lead to either too many small clusters or one large cluster with many points labeled as noise. Finding optimal values often requires domain knowledge or using techniques like the k-distance graph, which can be computationally intensive.

DBSCAN is better when clusters have irregular shapes and noise must be detected, while K-Means works best for well-separated, spherical clusters. Overall, DBSCAN is more flexible, but K-Means is simpler and faster for large, clean datasets.
